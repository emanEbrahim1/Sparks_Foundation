***segment wise (sales, Discount, Profit)***

```python
df['Segment'].value_counts()
```

```python
df_segment= df.groupby(['Segment'])[['Sales', 'Discount', 'Profit']].mean()
df_segment
```

```python
df_segment.plot.pie(subplots=True, 
                    autopct='%1.1f%%',
                    figsize=(18, 20),
                    shadow=True,
                    startangle = 90,
                    labels = df_segment.index)
plt.title('Segment wise analysis of Sale, Discount, profit')
```

***city wise Profit***

```python
df_city= df.groupby(['City'])[['Sales', 'Discount', 'Profit']].mean()
df_city = df_city.sort_values('Profit')
df_city.head()
```

```python
#Low 
df_city['Profit'].head(20).plot(kind='bar',figsize=(18,5))
plt.title('City wise analysis of Sale, Discount, profit')
```

```python
#High 
df_city['Profit'].tail(20).plot(kind='bar',figsize=(18,5))
plt.title('City wise analysis of Sale, Discount, profit')
```

### State wise (Sales , profit , discount)

```python
df['State'].value_counts().head(10)
```

```python
df_state= df.groupby(['State'])[['Sales', 'Discount', 'Profit']].mean()
df_state.head(10)
```

**[1] Statewise Sales Analysis**

```python
df_state['Sales'].plot(kind='pie',
                        figsize = (20,20),
                        autopct='%1.1f%%',
                        shadow=True)
plt.title('State wise analysis of Sale',fontsize=15)
```

- Highest amount of sales= ***Wyoming(11.8%)***
- Lowest amount of sales= ***South Dakota(0.8%)***

**[2] Statewise Profit Analysis**

```python
df_state1=df_state.sort_values('Profit')

df_state1[['Profit']].plot(kind = 'bar', figsize = (18,5))
plt.title('State wise Profit Analysis', fontsize = 15)
plt.ylabel('Profit per Sate')
plt.xlabel('States')
plt.show()
```

### RESULT

- Vermont: Highest Profit

- Ohio: Lowest Profit

**[3] Statewise Discount Analysis**

```python
df_state1['Discount'].plot(kind='bar',figsize=(18,5))
plt.title('State wise analysis of Discount', fontsize=15)
```

**Illinois at the top**

### Quantity wise (sales, profit , discount) 

```python
df_quantity = df.groupby(['Quantity'])[['Sales', 'Discount', 'Profit']].mean()
df_quantity.head(10)
```

```python
df_quantity.plot.pie(subplots=True, 
                    autopct='%1.1f%%',
                    figsize=(20, 20),
                     pctdistance=0.69,
                    startangle=90,
                    shadow=True,
                    labels = df_quantity.index)
plt.title('Quantity wise analysis of Sale, Discount, profit')
```

### Sub-Category wise (sales, profit , discount) 

```python
df_sub = df.groupby(['Sub-Category'])[['Sales', 'Discount', 'Profit']].mean()
df_sub.head(10)
```

**[1] Based on sales**

```python
plt.figure(figsize = (15,15))
plt.pie(df_sub['Sales'], labels = df_sub.index, autopct = '%1.1f%%')
plt.title('Sub-Category Wise Sales Analysis', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

**Copier then Machine have High sales.**

**[2] Based on profit**

```python
df_sub.sort_values('Profit')[['Sales','Profit']].plot(kind='bar',
                                                              figsize= (18,5),
                                                              label=['Avg Sales Price($)','Profit($)'])
```

**Copier : Highest profit as well as sell**

**[3] Based on discount**

```python
plt.figure(figsize = (15,15))
plt.pie(df_sub['Discount'], labels = df_sub.index, autopct = '%1.1f%%')
plt.title('Sub-Category Wise Discount Analysis', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

**Binders , Machines and then tables have high Discounts.**
