***Importing Required Libraries***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

***Load Dataset***

```python
df = pd.read_csv("globalterrorism.csv",encoding='latin1')
df.head()
```

```python
df.columns.values
```

**I will not use all columns so i will drop the columns that i won't use..**

```python
ds = df[['iyear' , 'imonth','iday','country_txt','provstate','region_txt','city','latitude','longitude',
         'attacktype1_txt','nkill','nwound','target1','summary','gname','targtype1_txt','weaptype1_txt','motive']]
```

```python
ds.rename(columns={'iyear':'Year','imonth':'Month','iday':"day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'latitude',
    'longitude':'longitude','summary':'summary','attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill',
     'nwound':'Wound'},inplace=True)
```

```python
ds.head()
```

```python
ds.shape
```

```python
ds.isnull().sum()
```

```python
ds['Wound'] = ds['Wound'].fillna(0)
ds['kill'] = ds['kill'].fillna(0)
```

```python
ds['Casualties'] = ds['kill'] + ds['Wound']
```

```python
ds.info()
```

```python
ds.describe()
```

```python
year = ds['Year'].unique()
years_count = ds['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = year,
           y = years_count,
           palette = "tab10")
plt.xticks(rotation = 50)
plt.xlabel('Attacking Year',fontsize=15)
plt.ylabel('Number of Attacks Each Year',fontsize=15)
plt.title('Attacks In Years',fontsize=25)
plt.show()
```

```python
pd.crosstab(ds.Year, ds.Region).plot(kind='area',stacked=False,figsize=(18,10))
plt.title('Terrorist Activities By Region In Each Year',fontsize=20)
plt.ylabel('Number of Attacks',fontsize=15)
plt.xlabel("Year",fontsize=15)
plt.show()
```

```python
attack = ds.Country.value_counts()[:10]
attack
```

```python
ds.Group.value_counts()[1:10]
```

```python
plt.subplots(figsize=(18,5))
sns.barplot(ds['Country'].value_counts()[:10].index,ds['Country'].value_counts()[:10].values)
plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.show()
```

```python
df1 = ds[['Year','kill']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(20,10))
df1.plot(kind='bar',alpha=0.7,ax=ax4)
plt.xticks(rotation = 90)
plt.title("People Died Due To Attack",fontsize=20)
plt.ylabel("Number of killed peope",fontsize=15)
plt.xlabel('Year',fontsize=15)
```

```python
ds['City'].value_counts().to_frame().sort_values('City',axis=0,ascending=False).head(15).plot(kind='bar',figsize=(18,5),color='blue')
plt.xticks(rotation = 45)
plt.xlabel("City",fontsize=15)
plt.ylabel("Number of attack",fontsize=15)
plt.title("Top 10 most effected city",fontsize=20)
plt.show()
```

```python
ds['Attacktype'].value_counts().plot(kind='bar',figsize=(18,5),color='pink')
plt.xticks(rotation = 45)
plt.xlabel("Attacktype",fontsize=15)
plt.ylabel("Number of attack",fontsize=15)
plt.title("Name of attacktype",fontsize=20)
plt.show()
```

```python
ds[['Attacktype','kill']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(18,5),color=['magenta'])
plt.xticks(rotation=45)
plt.title("Number of killed ",fontsize=20)
plt.ylabel('Number of people',fontsize=15)
plt.xlabel('Attack type',fontsize=15)
plt.show()
```

```python
ds[['Attacktype','Wound']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(18,5),color=['lightblue'])
plt.xticks(rotation=45)
plt.title("Number of wounded  ",fontsize=20)
plt.ylabel('Number of people',fontsize=15)
plt.xlabel('Attack type',fontsize=15)
plt.show()
```

```python
plt.subplots(figsize=(18,5))
sns.countplot(ds["Targettype"],order=ds['Targettype'].value_counts().index,palette="YlOrBr_r",edgecolor=sns.color_palette("Dark2"));
plt.xticks(rotation=90)
plt.xlabel("Attacktype",fontsize=15)
plt.ylabel("count",fontsize=15)
plt.title("Attack per year",fontsize=20)
plt.show()
```

```python
ds['Group'].value_counts().to_frame().drop('Unknown').head(15).plot(kind='bar',color='darkblue',figsize=(18,5))
plt.title("Top 10 terrorist group attack",fontsize=20)
plt.xlabel("terrorist group name",fontsize=15)
plt.ylabel("Attack number",fontsize=15)
plt.show()
```

```python
ds[['Group','kill']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values('kill',ascending=False).head(15).plot(kind='bar',color='red',figsize=(18,5))
plt.title("Top 10 terrorist group attack",fontsize=20)
plt.xlabel("terrorist group name",fontsize=15)
plt.ylabel("No of killed people",fontsize=15)
plt.show()
```
